```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Inputing the Data
data=pd.read_csv('Data.csv')
X=data.loc[:,'X']
Y=data.loc[:,'Y']
```

```python
#Plotting the Data on Graph
plt.figure(figsize=(10,6))
plt.scatter(X,Y)
plt.title("DataSet")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#Initialising hyperparameters
m = 0
c = 0
L = 0.00001  # The learning Rate
iterations = 1000  # The number of iterations to perform gradient descent
```

```python
#Using Batch Gradient
H1 = m*X +c
for i in range(iterations):
    H1=m*X +c
    m = m - L*sum((H1-Y)*X)
for i in range(iterations):
    H1=m*X +c
    c = c - L*sum((H1-Y))
H1 = m*X +c
```

```python
#Using scholastic Gradient Descent
m=0
c=0
H2 =m*X+c 
for i in range(iterations):
    for j in range(len(X)):
        H2[j]=m*X[j] +c
        m = m - L*(H2[j]-Y[j])*X[j]
    for j in range(len(X)):
        H2[j]=m*X[j] +c
        c = c - L*(H2[j]-Y[j])
H2 =m*X+c
```

```python
#Plotting the Lines
plt.figure(figsize=(15,9))
plt.scatter(X,Y)
plt.plot(X,H1,color="red",label = "Using Batch Gradient Descent")
plt.plot(X,H2,color ="black",label= "Using Scholastic Gradient Descent")
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
