```python
import numpy as np
from scipy.stats import mode
from sklearn.metrics import accuracy_score
```

```python
train = pd.read_csv("C:\\Users\\singh\\OneDrive\\Desktop\\Ye dil maange more\\pendigits_tra.csv")
test = pd.read_csv("C:\\Users\\singh\\OneDrive\\Desktop\\Ye dil maange more\\pendigits_tes (1).csv")
```

```python
x_train=train.iloc[:,:-1] 
y_train=train.iloc[:,-1]
x_test=test.iloc[:,:-1]
y_test=test.iloc[:,-1]
```

```python
class KNN:
    def __init__(self, k):
        self.k = k
    def load(self, dataset):
        [self.x,self.y]=dataset
    def eucledian(self,p1,p2):
        dist = np.sqrt(np.sum((p1-p2)**2))
        return dist     
    def predict(self, x_train, y, x_test, k):
        op_labels = []
        for item in x_test:
            arr = []
            point_dist = []
            for j in range(len(x_train)):
                arr = np.array(x_train[j,:])
                distances = self.eucledian(np.array(x_train[j,:]),item) 
                point_dist.append(distances) 
            point_dist = np.array(point_dist)
            dist = np.argsort(point_dist)[:k]
            labels = y[dist]
            lab = mode(labels) 
            lab = lab.mode[0]
            op_labels.append(lab)
        return op_labels
    def score(self,x_train,y_train,x_test,y_test,k):
        return accuracy_score(y_test,predict(x_train,y_train,x_test,k))
```

```python
model=KNN(1)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
model.load([x_train, y_train])
y_pred = model.predict(x_train,y_train,x_test,1)
```

```python
accuracy_score(y_test, y_pred)
```

To decide the hyperparameteres using the test accuracy is not the best way. Why?

The choice of hyperparameters based on the test accuracy will in most general case lead to overfitting which itself will make the model less effective against test data .Therefore it is not suggested and methods like cross-validation could be used instead
